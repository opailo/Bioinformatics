```python
def read_genome(fname):
  genome = ""
  with open(fname, 'r') as f:
    for line in f:
      if not line[0] == '>':
        genome += line.rstrip()
  
  return genome

genome = read_genome('lambda_virus.fa')

genome[:10]
```
